**Statistics Part 2**

1. What is hypothesis testing in statistics?
- Hypothesis sampling is a statatical method to determine whether there is enough evidance in a sample to support a particular claim about a population. It involves comparing a null hypothesis (H1) using a test statistic and a p-value.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- This null hypothesis assumes no effect or difference exists in the population acting as the default claim. The alternative hypothesis suggests a real effect or difference is present , challenging based on sample data evidence.

3. What is the significance level in hypothesis testing, and why is it important?
- Significance level (alpha) is the threshold probability (commanly set to 0.05 ) used in hypothesis. It helps control the risk of incorrectly rejecting (H0) ensuring decisions are statistically reliable.

4. What does a P-value represent in hypothesis testing?
- P value represents the probability of obtaining results as extreme as the observed data assuming the null hypothesis is true. A smaller P value (<alpha) suggest stronger evidence against H0, leading to its rejection.

5.  How do you interpret the P-value in hypothesis testing?
- Interpreting the P value depends on its comparisions with the significance level(alpa):
- If P value less than alpha > there is strong evidence against the null hypothesis , so we reject H0.

6. What are Type 1 and Type 2 errors in hypothesis testing?
- In hypothesis testing:
- Type 1 Error (False Positive): Incorrectly rejecting the null hypothesis (H₀) when it's actually true. This means detecting an effect that doesn't exist.
- Type 2 Error (False Negative): Failing to reject H₀ when the alternative hypothesis (H₁) is true. This means missing a real effect that actually exists.
Balancing these errors is crucial in statistical decision-making!

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- A one-tailed test checks for an effect in one specific direction (either greater or smaller), while a two-tailed test checks for an effect in  both directions (greater and smaller). A two-tailed test is more conservative since it considers both possibilities.

8. What is a confidence interval, and how is it used to interpret statistical results?  
- A confidence interval provides a range around an estimate that likely contains the true population parameter, indicating result reliability.

9. What is the margin of error, and how does it affect the confidence interval?  
- The margin of error represents uncertainty; a larger margin widens the confidence interval, reducing precision.

10. How is Bayes' Theorem used in statistics, and what is its significance?  
- Bayes' Theorem updates probabilities based on prior knowledge and new evidence, crucial for decision-making and inference.

11. What is the Chi-square distribution, and when is it used?  
- The Chi-square distribution is used for hypothesis testing in categorical data, including goodness-of-fit and independence tests.

12. What is the Chi-square goodness of fit test, and how is it applied?  
- It checks if observed categorical data matches expected distributions, commonly used in genetics and survey analysis.

13. What is the F-distribution, and when is it used in hypothesis testing?  
- The F-distribution is applied in variance analysis, such as ANOVA, to compare multiple sample variances.

14. What is an ANOVA test, and what are its assumptions?  
- ANOVA compares means across multiple groups; it assumes normality, equal variances, and independent observations.

15. What are the different types of ANOVA tests?  
- One-way ANOVA (single factor), two-way ANOVA (two factors), and repeated measures ANOVA (same subjects over time).

16. What is the F-test, and how does it relate to hypothesis testing?  
- The F-test compares variances between groups, commonly used in ANOVA and regression analysis.

17. What is a Type I error in hypothesis testing, and what are its consequences?  
- A Type I error occurs when a true null hypothesis is incorrectly rejected, leading to false-positive results.

18. What is a Type II error in hypothesis testing, and how does it differ from Type I error?  
- A Type II error happens when a false null hypothesis is incorrectly accepted, causing false-negative results.

19. What is statistical power, and why is it important in hypothesis testing?  
- Statistical power measures the probability of correctly rejecting a false null hypothesis; higher power reduces Type II errors.

20. What is the p-value, and how is it used in hypothesis testing?  
- The p-value quantifies evidence against the null hypothesis; a lower p-value suggests stronger significance.

21. What is the difference between one-tailed and two-tailed tests in hypothesis testing?  
- One-tailed tests assess deviations in a single direction, while two-tailed tests examine both directions.

22. What is regression analysis, and how is it used in statistics?  
- Regression analysis models relationships between variables, helping predict outcomes based on input data.

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
from scipy.stats import norm

data = np.random.normal(50, 10, 100)
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
alpha = 0.05
z_score = norm.ppf(1 - alpha/2)

margin_of_error = z_score * (sample_std / np.sqrt(len(data)))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"95% Confidence Interval: {confidence_interval}")
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

sample_mean, population_mean, std_dev, n = 52, 50, 5, 30
z_score = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p_value = norm.sf(abs(z_score)) * 2

x = np.linspace(-3, 3, 100)
y = norm.pdf(x)
plt.plot(x, y)

plt.axvline(-1.96, color='r', linestyle='dashed', label="Critical Value (-1.96)")
plt.axvline(1.96, color='r', linestyle='dashed', label="Critical Value (1.96)")
plt.axvline(z_score, color='b', linestyle='dashed', label=f"Z-score ({z_score:.2f})")

plt.legend()
plt.title("Decision Region for Two-Tailed Z-Test")
plt.show()
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(alpha=0.05, beta=0.2):
    x = np.linspace(-3, 3, 100)
    y = norm.pdf(x)
    
    plt.plot(x, y, label="Distribution")
    plt.axvline(-norm.ppf(1 - alpha), color='r', linestyle='dashed', label="Type I Error")
    plt.axvline(norm.ppf(beta), color='g', linestyle='dashed', label="Type II Error")
    
    plt.legend()
    plt.title("Type I and Type II Errors")
    plt.show()

visualize_errors()
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
from scipy.stats import t

def confidence_interval(data, alpha=0.05):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)
    t_score = t.ppf(1 - alpha/2, df=n-1)
    
    margin_of_error = t_score * (sample_std / np.sqrt(n))
    return sample_mean - margin_of_error, sample_mean + margin_of_error

data = np.random.normal(50, 10, 100)
ci = confidence_interval(data)
print(f"95% Confidence Interval: {ci}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior, likelihood, marginal_likelihood):
    return (likelihood * prior) / marginal_likelihood

prior = 0.5  # Initial belief
likelihood = 0.8  # Probability of data given the hypothesis
marginal_likelihood = 0.7  # Overall probability of observing the data

posterior = bayes_theorem(prior, likelihood, marginal_likelihood)
print(f"Updated probability using Bayes' Theorem: {posterior}")
```
